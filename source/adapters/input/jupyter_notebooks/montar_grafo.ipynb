```python
import matplotlib.pyplot as plt
import networkx as nx

# Initialize the graph
G = nx.DiGraph()

# Add nodes
## Central node
G.add_node("St. Maximus the Confessor")

## Major Sub-Nodes
G.add_nodes_from(["Theological Context", "Key Texts", "Heresies Contested"])

## Additional Nodes
G.add_nodes_from(["Byzantine Empire", "Christological debates", "Monasticism",
                  "Ambigua", "Opuscula theologica et polemica", "Mystagogia",
                  "Monothelitism", "Monophysitism", "Origenism"])

# Add edges
## Connecting major sub-nodes to central node
G.add_edges_from([("St. Maximus the Confessor", "Theological Context"),
                  ("St. Maximus the Confessor", "Key Texts"),
                  ("St. Maximus the Confessor", "Heresies Contested")])

## Connecting additional nodes to major sub-nodes
G.add_edges_from([("Theological Context", "Byzantine Empire"),
                  ("Theological Context", "Christological debates"),
                  ("Theological Context", "Monasticism"),
                  ("Key Texts", "Ambigua"),
                  ("Key Texts", "Opuscula theologica et polemica"),
                  ("Key Texts", "Mystagogia"),
                  ("Heresies Contested", "Monothelitism"),
                  ("Heresies Contested", "Monophysitism"),
                  ("Heresies Contested", "Origenism")])

# Draw the graph
pos = nx.spring_layout(G)  # positions for all nodes

nx.draw(G, pos, with_labels=True, node_color="lightblue", font_weight="bold", node_size=2000, font_size=10)

plt.title("Ideas Combated as Heresies by St. Maximus the Confessor")
plt.show()
```
